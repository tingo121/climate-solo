The GCPT catalogues every operating coal-fired generating unit, every new unit proposed since 2010, and every unit retired since 2000. Units often consist of a boiler and turbine, and several units may make up one coal-fired power station. The map and underlying data is updated bi-annually, in January and July. ### Emissions from Bunker Fuels

"Heavy Fuel Oil (HFO) is a category of fuel oils of a tar-like consistency. Also known as bunker fuel, or residual fuel oil, HFO is the result or remnant from the distillation and cracking process of petroleum." This is the type of oil often used on ships and has caused great damage in the past due to oil spills. 

```python
#look at bunker fuels
data.loc[data['Series Name'] == 'CO2 emissions by sector (Mt CO2 eq) - Bunker Fuels']
```

```python
#filtering down a smaller dataframe with just my desired info
bunker_fuels_co2 = data[data['Series Name'] == 'CO2 emissions by sector (Mt CO2 eq) - Bunker Fuels']
bunker_fuels_co2 = bunker_fuels_co2.drop(['Series Name', 'Series Code', 'Country Code', '2019', '2020', '2021', '2030', '2050'], axis = 1)
bunker_fuels_co2.head(3)
```

```python
#need to change to long data instead of wide data before charting years?
bunker_fuels_co2 = pd.melt(bunker_fuels_co2, 
                           id_vars = 'Country Name', 
                           value_vars = ['1990', '2000', '2010', '2015', '2016', '2017', '2018'],
                           var_name='Year', 
                           value_name='Value'
                          ).sort_values(['Country Name', 'Year']).reset_index(drop=True)
bunker_fuels_co2
```

```python
sns.set(rc={"figure.figsize":(10, 6)})

bunker_fuels_co2_lineplot = \
sns.lineplot(
    data = bunker_fuels_co2,
    x = bunker_fuels_co2['Year'],
    y = bunker_fuels_co2['Value'],
    hue = bunker_fuels_co2['Country Name'],
    legend = 'auto',
).set(title = 'Bunker Fuel CO2 Emissions over Time', 
      xlabel = 'Year', 
      ylabel = 'Metric Tons of CO2')

sns.move_legend(plt.gca(), loc="upper left", bbox_to_anchor=(1, 1))
```

```python
bunker_fuels_total = data[data['Series Name'] == 'Total GHG emissions by sector (Mt CO2 eq) - Bunker Fuels']
bunker_fuels_total = bunker_fuels_total.drop(['Series Name', 'Series Code', 'Country Code', '2019', '2020', '2021', '2030', '2050'], axis = 1)
bunker_fuels_total.head(3)
```

```python
#need to change to long data instead of wide data before charting years?
bunker_fuels_total = pd.melt(bunker_fuels_total, 
                           id_vars = 'Country Name', 
                           value_vars = ['1990', '2000', '2010', '2015', '2016', '2017', '2018'],
                           var_name='Year', 
                           value_name='Value'
                          ).sort_values(['Country Name', 'Year']).reset_index(drop=True)
bunker_fuels_total
```

```python
sns.set(rc={"figure.figsize":(10, 6)})

bunker_fuels_total_lineplot = \
sns.lineplot(
    data = bunker_fuels_total,
    x = bunker_fuels_total['Year'],
    y = bunker_fuels_total['Value'],
    hue = bunker_fuels_total['Country Name'],
    legend = 'auto',
).set(title = 'Bunker Fuel Total GhG Emissions over Time', 
      xlabel = 'Year', 
      ylabel = 'Metric Tons of CO2 Equivalent')

sns.move_legend(plt.gca(), loc="upper left", bbox_to_anchor=(1, 1))
```

At a glance, the total bunker fuels GhG emissions and the CO2 bunker fuels emissions look very similar indeed. Do they even differ at all? Let's compare.

We can look at this ourselves and see that the numbers are identical. Or, we can code a simple comparison with our two long form dataframes we made in an earlier step. 

```python
bunker_fuels_co2.equals(bunker_fuels_total)
```

Assuming our data is complete, this suggests that bunker fuels emit CO2 rather than other GhG. Because they are equal, we can delete the data from one of these series from our original dataframe. 

```python
data = data.drop(data[data['Series Name'] == 'CO2 emissions by sector (Mt CO2 eq) - Bunker Fuels'].index)
```

The data shows us that most countries in our list have a fairly steady level of emissions from bunker fuels. China and the United States are by far the highest emitters from 2000-2018 onward. China and Russia both dipped in their bunker fuel emissions levels between 1990 and 2000, but whereas Russia's leveled off, China's levels surged upwards again.
